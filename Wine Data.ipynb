```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
df.columns
```

```python
X = df[['fixed_acidity', 'volatile_acidity', 'citric_acid', 'residual_sugar',
       'chlorides', 'free_sulfur_dioxide', 'total_sulfur_dioxide', 'density',
       'pH', 'sulphates', 'alcohol']]
y = df['quality']
```

```python
from sklearn.model_selection import train_test_split     #for doing the split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 101) 
score = {}
```

```python
#LINEAR REGRESSION

from sklearn import datasets, linear_model
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import statsmodels.api as sm
from scipy import stats

#X = sm.add_constant(X_test) # adding a constant

#model = sm.OLS(y, X).fit()
#predictions = model.predict(X) 



linear_regressor = LinearRegression()
linear_regressor.fit(X_train, y_train)
y_pred = linear_regressor.predict(X_test)
print(linear_regressor.coef_)


```

```python
linreg = LinearRegression()
linreg.fit(X_train, y_train)
```

```python
y_pred = linreg.predict(X_test)  
print(metrics.mean_absolute_error(y_test, y_pred))
print(metrics.mean_squared_error(y_test, y_pred))
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
accuracy = linreg.score(X_test, y_test)
score.update({'Linear Regression': accuracy*100})
print(accuracy)
```

```python
#DECISION TREE CLASSIFIER

from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)
```

```python
predictions = dtree.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix


print(classification_report(y_test, predictions))

```

```python
from sklearn.metrics import accuracy_score

accuracy_dt = accuracy_score(y_test, predictions)
score.update({'Decision Tree Classifier': accuracy_dt*100})
print(accuracy_dt)
```

```python
#RANDOM FOREST CLASSIFIER

rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train, y_train)
predicted_rfc = rfc.predict(X_test)
print(classification_report(y_test, predicted_rfc))
```

```python
rfc_acc_score = accuracy_score(y_test, predicted_rfc)
score.update({'Random Forest Classifier': rfc_acc_score*100})
print(rfc_acc_score)
```

```python
#GAUSSIAN NAIVE BAYES

from sklearn.naive_bayes import GaussianNB
gb = GaussianNB()
gb.fit(X_train,y_train)
predicted_gb = gb.predict(X_test)
print(classification_report(y_test, predicted_gb))

```

```python
gb_acc_score = accuracy_score(y_test, predicted_gb)
score.update({'Gaussian Naive Bayes': gb_acc_score*100})
print(gb_acc_score)
```

```python
model_acc = pd.DataFrame()
model_acc['Models'] = score.keys() 
model_acc['Accuracy'] = score.values()
model_acc
```

```python
from matplotlib.pyplot import xticks
sns.lineplot(x='Models', y='Accuracy',data=model_acc)
xticks(rotation=90)
```
